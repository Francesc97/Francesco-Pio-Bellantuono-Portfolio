```python
#decollete
nomi = []
prezzi_iniziali = []
prezzi_scontati = []
categorie = []
brands=[]
taglie = []

for pagina in range(1, 15):
    url = "https://www.melluso.com/it/melluso-donna/decollete.html?p={pagina}"+"&price=52.00-162.00"
    print("Elaboro pagina", pagina, "alla URL:", url)

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
#estraggo le categorie
    categoria_tag = soup.find("span", class_="base")
    categoria = categoria_tag.text.strip() if categoria_tag else "N/A"
#estraggo il nome de brand
    brand_tag = soup.find("img", {"alt": True})
    brand = brand_tag.get("alt", "").split(":")[0].strip()

    # Estrazione taglie (filtro label con 'taglia' in 'for')
    taglie_divs = soup.find_all("div", class_="am-shopby-multiselect-swatch")
    taglie_pagina = [div.get("data-option-label") for div in taglie_divs]
    taglie_numeriche = list(map(float, taglie_pagina)) if taglie_pagina else []
    taglia_min = min(taglie_numeriche)
    taglia_max = max(taglie_numeriche)

    prodotti = soup.find_all('li', class_='product-item')
    
    for prodotto in prodotti:
#estraggo il nome del prodotto
        nome = prodotto.find("a", class_="product-item-link")
        nome = nome.text
        nomi.append(nome)
#estraggo e sistemo il prezzo

        sconto = prodotto.find("span", attrs={"data-price-type": "finalPrice"})
        sconto = sconto.text
        sconto = sconto[:-2]
        sconto = sconto.replace(",", ".")
        sconto = float(sconto)
        
        prezzo = prodotto.find("span", attrs={"data-price-type": "oldPrice"})
        prezzo = prezzo.text
        prezzo = prezzo[:-2]
        prezzo = prezzo.replace(",", ".")
        prezzo = float(prezzo)
#aggiungo tutto nelle apposite liste 
        prezzi_scontati.append(sconto)
        prezzi_iniziali.append(prezzo)
        categorie.append(categoria)
        brands.append(brand)
        taglie.append((taglia_min, taglia_max))

df_dec = pd.DataFrame({
    "nome": nomi,
    "prezzo_iniziale": prezzi_iniziali,
    "prezzo_scontato": prezzi_scontati,
    "categoria": categorie,
    "brand": brands,
    "taglia_min": [t[0] for t in taglie],
    "taglia_max": [t[1] for t in taglie]
})

df_dec 
```

```python
#proviamo ad estrarre i colori dai nomi dei prodotti
colori = ["nero", "bianco", "nude","blu denim","cognac", "blu","beige kiss", "platino", "cielo","coloniale",
          "fondotinta", "verde", "jeans", "argento", "beige", "macchiato", "greige", "abyss","notte", "phard"]
colori_trovati = []  # lista per salvare i colori trovati

for nome in df_melluso["nome"]:
    colore_trovato = "sconosciuto"  # valore di default
    for colore in colori:
        if colore in nome:
            colore_trovato = colore
          
    colori_trovati.append(colore_trovato)

df_melluso["colore"] = colori_trovati
df_melluso
```
